## 8. Хитрый вопрос

Правда ли, что при более сильном ветре человеку ощущается, что на улице холодне, чем на самом деле? Нарисуйте график, подтверждающий или опровергающий эту гипотезу. Обоснуйте. Возможно, вам придется создать новый признак.

Рассуждение: Заведем новую колонку dT в которой будет разница между ощущаемой и настоящей температурой. Если разность будет отрицательной, значит человеку кажется, что на улице холоднее. Возмем по вертикальной оси нашу колонку, а по горизонтальной скорость ветра и увидим, что отрицательных значений немного, значит человеку в зависимости от сильного ветра не кажется что на улице холоднее.

```python
df['dT'] = df['atemp'] - df['temp']
sns.pointplot(x='windspeed(ms)', y='dT', data=df)
```
